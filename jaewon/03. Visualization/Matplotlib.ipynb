## 범례 표시하기

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0,3,.02)
c = np.exp(a)
d = c[::-1]

plt.plot(a, c, 'r--', label='Model length')
plt.plot(a, d, 'b:', label = 'Data length')
plt.plot(a, c+d, 'g', label='Total message length')

plt.legend(loc='best', shadow = True)
plt.show()
```

## 서브플롯 그리기
- subplot() 함수를 이용해서 하나의 그림에 여러 개의 그래프를 그릴 수 있다
    -nrow, ncols, index의 인수를 이용해 서브플롯의 위치를 설정한다

```python
x1 = np.linspace(0, 5, 50)
x2 = np.linspace(0, 2, 50)

y1 = np.cos(2*np.pi*x1)+np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1) #nrow = 2 ncol=1 index =1
plt.plot(x1, y1,'o--')
plt.title('1st gtaph')
plt.xlabel('time(s)')
plt.ylabel('Damped')

plt.subplot(2,1,2)
plt.plot(x2, y2,'.-')
plt.title('2nd gtaph')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

#plt.tight_layout()  #서브플롯의 레이아웃을 자동으로 맞춰준다
plt.show()
```

### subplot() 함수
- Figure, Axes 객체를 반환한다
- Axes 클래스: 공간이 확보된 이미지 영역 정보를 가지고 있다
      - 주요 인자
      >- nrows, ncols: 서프플롯의 행과 열 크기 지정
      >- sharex, sharey: boolean값 (default : False), x, y축의 값 공유 여부 

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, sharey = True)

fig.set_size_inches((8,8))  # 튜플 입력을 위해 가로 2개 사용

axes[0][0].scatter(np.random.random(100), np.random.random(100), c='cyan', edgecolor = 'red', alpha = 0.5)
axes[0][1].scatter(np.random.random(100), np.random.random(100), c='magenta', edgecolor = 'green', alpha = 0.5)
axes[1][0].scatter(np.random.random(100), np.random.random(100), c='yellow', edgecolor = 'blue', alpha = 0.5)
axes[1][1].scatter(np.random.random(100), np.random.random(100), c='white', edgecolor = 'black', alpha = 0.5)

axes[0][0].set_title('sibplot-0')
axes[0][1].set_title('sibplot-1')
axes[1][0].set_title('sibplot-2')
axes[1][1].set_title('sibplot-3')

plt.show()
```

# 막대 그래프

## 수직 막대 그래프

```python
x = np.arange(0,20)
y = x*2

x_ticks = np.arange(len(x)) # x축의 눈금을 표시할 index 
plt.bar(x,y,label = 'Blue bar', color = 'blue', edgecolor = 'yellow', linewidth = 1)
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')
plt.xticks(x_ticks, x_ticks+1 )
plt.legend()
plt.show()
```

## 수평 막대 그래프

```python
x = np.arange(0,20)
y = x*2

y_ticks = np.arange(len(x)) # x축의 눈금을 표시할 index 

# 수평 막대 그래프를 그린다고 해서 x와 y의 위치가 바뀌는 것은 아니다
plt.barh(x,y,label = 'Blue bar', color = 'blue', edgecolor = 'yellow', linewidth = 1)
plt.ylabel('bar number')
plt.xlabel('bar height')
plt.title('Bar Chart Example')
plt.yticks(y_ticks, y_ticks+1 )
plt.legend(loc = 'lower right')
plt.show()
```

```python
label = ['Seoul', 'Busan', 'Incheon', 'Gwanji', 'Jeju']
men_means = [20, 34, 30, 85, 27]
women_means = [25, 32, 34, 56, 20]

x_ticks_index = np.arange(len(men_means))
width = 0.35 # 막대 그래프 폭의 크기
plt.bar(x_ticks_index-width/2, men_means, width = width, color = 'dodgerblue', label = 'Men')# x-width /2 -> 두 막대 값의 중간점을 찾는 코드
plt.bar(x_ticks_index-width/2, women_means, width = width, color = 'violet', label = 'Women')

plt.xlabel('city')
plt.ylabel('Score')
plt.title('score by group and gender')
plt.xticks(x_ticks_index, label)
plt.legend()
plt.show()
```

# 산점도 그리기
- 산점도(scatter plot)는 직교좌표계를 이용해 두 변수 간의 상관관계를 확인하고자 할 때 사용하는 플롯이다

```python
x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 5]

y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker = 'v', color = 'green')
plt.scatter(x2, y3, marker = '^', color = 'red')
plt.title('Scatter Plot Example')
plt.show()
```

```python
for color in ['blue', 'orange', 'green']:
    n = 50
    x, y = np.random.rand(2, n)  # [0,1] (0이상 1 미만) 사이의 난수를 2x50 모양으로 생성한다 -> 이차원
    scale = 190.0 * np.random.rand(n) + 10 # [10, 200)  -> 10 이상 200 미만 사이의 난수 50개를 발생한다
    plt.scatter(x, y, c=color, s=scale, label =color, alpha = 0.3, edgecolors = "none")
plt.legend(loc = (1.04, 0.7))
plt.grid(True) #그래프에 좌표를 그려주는 함수
plt.show()
```

# 원 그래프 그리기

```python
import matplotlib.pyplot as plt

label = ['python', 'JAVA', 'C']
# autopct: 각 조각 위에 비율(%)을 표시하는 형식 문자열을 지정한다 -> 문자열에서 formated 스트링과 유사
plt.pie([0.3, 0.35, 0.4], labels=label, autopct='%.1f%%')
plt.title('Popular language ranking')
plt.legend(loc = (1, 0.3))
plt.show()
```

```python
data = [0.3, 0.35, 0.4]
label = ['python', 'JAVA', 'C']

colors = ['coral', 'dodgerblue', 'yellowgreen']

plt.pie(
    data, # 조각의 크기를 결정하는 값
    labels = label, # 조각의 이름 표시
    explode = [0.1, 0, 0], # 조각을 중심으로부터 얼마나 떨어지게 할 것인지를 설정한 [0,1: 반지름의 10%만큼 벗어남
    colors = colors, # 조각의 색깔
    autopct = '%.2f%%', # 비율을 표시하는 형식 문자열
    shadow = True, #그림자 표시 여부
    startangle = 90, # 조각의 시작 부분을 각도로 표시한다. 해당 각도에서부터 반시계 방향으로 회전하면서 조각을 표시 [기본값:0, 3시 방향]
    counterclock = False, # 반시계 방향으로 순서대로 표시할 지 결정 (기본값: True)
    rotatelabels = True, #조각의 이름을 회전한다. (기본값: False)

)
plt.title('Popular language Ranking')
plt.show()
```

# 박스플룻 그리기
- 박스플룻은 데이터 집합의 범위와 중앙값 등 통계 수치값을 확인하고자 하는 목적으로 사용된다
- 이상치(outlier)가 있는지 검사하기 수월하다
- 최소값: 제 1사분위에서 1.5IQR(inter quantile range:사분위수범위, Q3-Q1)을 뺀 위치
- 제 1사분위(Q1): 중앙값 기준으로 하위 50% 중의 중앙값, 전체 데이터 중 하위 25%에 해당하는 값이다
- 제 2사분위(Q2): 전체 데이터의 중앙값. 즉 50%의 위치에 중앙값(dedian)을 의미
- 제 3사분위(Q3): 중앙값 기준으로 상위 50% 중의 중앙값, 전체 데이터 중 상위 25%에 해당하는 값
- 최대값: 제 3사분위에서 1.5IQR을 더한 위치
- 최대값과 최소값 범위를 벗어나는 값을 이상치 데이터로 간주한다. 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a =np.random.normal(0,2.0,1000) # 평균이 0이고 표준 편차가 2인 1000개의 데이터를 만들어라
data_b = np.random.normal(-3.0, 1.5, 500) #평균이 -3이고 표준 편차가 1.5인 500개의 데이터
data_c = np.random.normal(1.2, 1.5, 1500)

plt.boxplot([data_a, data_b, data_c])
plt.ylim(-10.0, 10.0)
plt.xlabel('Data type')
plt.ylabel('Value')
plt.show()
```

```python
data_a =np.random.normal(0,2.0,1000)
print(f'min: {np.min(data_a):.2f}, median: {np.median(data_a):.2f}, max: {np.max(data_a):.2f}')
quantile = np.quantile(data_a, [0, 0.25, 0.5, 0.75, 1])
print(quantile)
IQR = quantile[3] - quantile[1]
print(IQR)
print('Min:',quantile[1]-(1.5*IQR))
print('Max:',quantile[3]+(1.5*IQR))
```

```python
# 이상치 데이터 확인
np.random.seed(0)

spread = np.random.rand(50) * 100 # 0이상 100 미만의 데이터 50개를 만들어 낸다
center = np.ones(25)*50 #50 중앙값 25개를 만든다
flier_high = np.random.rand(10) * 100 + 100 # 100 이상 200 미만의 데이터를 10개 만든다
flier_low = np.random.rand(10) * -100 # -99에서 0 사이의 데이터 10개를 만든다

data = np.concatenate([spread, center, flier_high, flier_low])
print(data.shape)

plt.boxplot(data, 1, 'rx')
plt.grid(True)
plt.title('Box plot')
plt.show()
```

```python
import random
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] =  12
matplotlib.rcParams['axes.unicode_minus'] = False #유니코드가 -기호로 설정되는 폰트가 깨지는 것을 방지

random.seed(10)

x_invest= list(range(5,46,5))

y_tv = [random.randint(1,64) for i in range(9)]
y_tv.sort()
y_google =[random.randint(1,74) for i in range(9)]
y_google.sort()

plt.plot(x_invest, y_tv, 'bs-', label='TV')
plt.plot(x_invest, y_google, 'ro--', label='Google')
plt.title('매체별 광고 성과 비교')
plt.xlabel('광고 투자 금액')
plt.ylabel('매출')
plt.legend()
plt.show()
```
